**Loading Data from file and normalization operation**

1. Min-Max Method

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/cancer.csv')
df
```

```python
column_name=df.columns.tolist()
print(column_name)
```

```python
min_range=0
max_range=1
min_value=df['population'].min()
max_value=df['population'].max()
df['population']=(df['population']-min_value)/(max_value-min_value)*(max_range-min_range)+min_range
print(df)
```

*2. Z Score method*

```python
df1=pd.read_csv('/content/housing-data.csv')
df1
```

```python
column_name1=df1.columns.tolist()
print(column_name1)
```

```python
mean=df1['HOUSTNSA'].mean()
std=df1['HOUSTNSA'].std()
df1['HOUSTNSA']=(df1['HOUSTNSA']-mean)/std
print(df1)
```
