Joining NumPy Arrays

```python
import numpy as np
a=np.array([1,2,3])
b=np.array([4,5,6])
c=np.vstack((a,b))
c
#print(c)
```

```python
import numpy as np
a=np.array([1,2,3,4])
b=np.array([5,6,7,8])
c=np.hstack((a,b))
c
```

```python
import numpy as np
a=np.array([1,2,3,4,5])
b=np.array([6,7,8,9,10])
c=np.column_stack((a,b))
c
```


NumPy Intersection and Difference

```python
import numpy as np
a=np.array([1,2,3,4])
b=np.array([3,4,5,6])
np.intersect1d(a,b)
```

```python
np.setdiff1d(a,b)
np.setdiff1d(b,a)
```

NumPy Mathematics

```python
import numpy as np
a=np.array([1,2,3,4])
b=np.array([3,4,5,6])
np.sum([a,b])
```

```python
np.sum([a,b],axis=0) #Addition of individual columns
```

```python
np.sum([a,b],axis=1) #Addition of individual rows
```

```python
import numpy as np
a=np.array([1,2,3,4])
b=np.array([5,6,7,8])
c=np.array([1,2,3,4])
d=np.array([5,6,7,8])
a=a+1
b=b*2
c=c-1
d=d/2
print("a+1=",a)
print("b*2=",b)
print("c-1=",c)
print("d/2=",d)
```

NumPy Mean, Median and Standard Deviation

```python
import numpy as np
a=np.array([10,48,23,9,45,61 ])
print("Standard Deviation ",np.std(a))
print("Mean ",np.mean(a))
print("Median ",np.median(a))
```
