```python
import math
a=19
b=7
c=5
```

Difference of 1st, 2nd and 3rd number

```python
print("Difference between",a,b,c,"is: ")
diff=a-b-c
print(diff)
```

Product of 1st and 2nd number

```python
print("Product of",a,b,"is ")
pro=a*b
print(pro)
```

Division of 1st and 2nd number

```python
print("\nQuotient of",a,b,"is: ")
print(a/b)
print("\nRemainder of",a,b,"is: ")
print(a%b)
```

Squaring of 1st,2nd and 3rd number

```python
print("Square of",a,b,c,"is: ")
print(a**2)
print(b**2)
print(c**2)
```

Square root of 1st,2nd and 3rd number

```python
print("/Square root of",a,b,c,"is: ")
print(math.sqrt(a))
print(math.sqrt(b))
print(math.sqrt(c))
```
