```python
def Humpty_Dumpty(a):
  if a%3==0 and a%5==0:
    return "Humpty_Dumpty"
  elif a%3==0:
    return "Humpty"
  elif a%5==0:
    return "Dumpty"
  else:
    return a
a=int(input("Enter number: "))
b=Humpty_Dumpty(a)
print(b)
```

```python
def sum(a):
  count=0
  for i in range(0,a+1):
    if i%3==0 or i%5==0:
      print(i,end=" ")
      count+=i
  print("")
  return count
a=int(input("Enter a number:"))
b=sum(a)
print(b)
```

```python
a=[[0,0],[0,0],[0,0]]
b=[[0,0,0],[0,0,0]]
c=[[0,0,0],[0,0,0],[0,0,0]]
print("Enter 3x2 matrix")
for row in range(3):
  for column in range(2):
    n=int(input("Enter number:"))
    a[row][column]=n
print("Enter 2x3 matrix")
for row in range(2):
  for column in range(3):
    n=int(input("Enter number:"))
    b[row][column]=n
for i in range(3):
  for j in range(3):
    for k in range(2):
      c[i][j]=c[i][j]+a[i][k]*b[k][j]
print("Result: ")
for i in range(3):
  for j in range(3):
    print(c[i][j],end=" ")
  print("")
```
