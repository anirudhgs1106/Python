```python
# Lottery Game
import random
import csv
import matplotlib.pyplot as plt
user=0
system=0
result=[]
n=int(input("Enter number of rounds: "))
for i in range(n):
  a=random.randint(1,10)
  b=random.randint(1,10)
  if a==b:
    winner="Draw"
  elif a>b:
    system+=1
    winner="System"
  else:
    user+=1
    winner="User"
  result.append({"User":b,"System":a,"Result":winner})
df=pd.DataFrame(result)
df.to_excel("results.xlsx",index=False)
user_rate=user/n * 100
system_rate=system/n * 100
plt.bar(['User','System'],[user_rate,system_rate],color=['Green','Red'])
plt.ylabel('Success Rate')
plt.xlabel('Player')
plt.show()
```
