Arithmetic Operators

```python
a=5
b=3
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a//b)
print(a%b)
print(a**b)
```

Logical Operators

```python
print(a and b)
print(a or b)
print(not a)
print(not b)
```

Comparision Operator

```python
print(a>b)
print(a<b)
print(a==b)
print(a>=b)
print(a<=b)
print(a!=b)
```

Bitwise Operator

```python
print(a&b)
print(a|b)
print(a^b)
print(a<<1)
print(b<<2)
print(a>>2)
print(b>>1)
```

Assignment Operator

```python
a=5
b=3
b+=3
print("b+=3 ",b)
a-=3
print("a-=3 ",a)
a*=2
print("a*=4 ",a)
b/=2
print("b/=2 ",b)
c=10
c%=(a+b)
print("c%=(a+b) ",c)
c//=3
print("c//=3 ",c)
a**=2
print("a**=2 ",a)
d=10
d&=2
print("d&=2 ",d)
d|=4
print("d|=4 ",d)
d>>=2
print("d>>=2 ",d)
d<<=2
print("d<<=2 ",d)
```

Identity and Membership Operator

```python
x=["Ani","Anirudh"]
y=["Anirudh","Ani"]
z=x
print(z is x)
print(z is y)
print(z is not y)
print("Ani" in y)
print("Anirudh" not in x)
```
