```python
# Rock, Paper and Scissor
import random
list1=["rock","paper","scissor"]
user=input("Enter rock,paper or scissor ")
if user!="rock" and user!="paper" and user!="scissor":
  print("Invalid choice")
  exit
system=random.choice(list1)
print(user)
print(system)

if user==system:
  print("Tie")

elif user=="rock" and system=="paper":
  print("Winner : User")
elif user=="paper" and system=="rock":
  print("Winner : System")

elif user=="rock" and system=="scissor":
  print("Winner : User")
elif user=="scissor" and system=="rock":
  print("Winner : System")

elif user=="scissor" and system=="paper":
  print("Winner : User")
else:
  print("Winner : System")
```
