```python
a=int(input("Enter a number: "))
i=0
c="*"
while i<=a:
  print(c*i)
  i+=1
```

```python
a=int(input("Enter a number: "))
i=a
c="*"
while i>=0:
  print(c*i)
  i-=1
```

```python
a=int(input("Enter a number: "))
c="*"
for i in range(a):
  print(c*(i+1))
```
