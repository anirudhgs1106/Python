```python
a=[]
for i in range(10):
  b=int(input("Enter number: "))
  a.append(b)
print("Even: ")
for i in range(10):
  if a[i]%2==0:
    print(a[i])
print("Odd: ")
for i in range(10):
  if a[i]%2!=0:
    print(a[i])
```
